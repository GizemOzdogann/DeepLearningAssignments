**Label encoding** ile kategorik verileri algoritmaların işleyebileceği sayısal forma dönüştürebiliriz.

```
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```
for column in mushroom_data.columns:
  mushroom_data[column] = le.fit_transform(mushroom_data[column])
```

```
mushroom_data.head()
```

Örneğin 'class' sütunundaki iki farklı veriyi p = 1, e = 0 olacak şekilde eşleştirerek numerik olarak dönüşüm sağladı. Bu sayede üzerinde işlem yapılabilecek bir formata sahip oldu.

```
X = mushroom_data.drop(columns = "class")
```

```
y = mushroom_data['class']
```

```
X.head(2)
```

```
y.head(2)
```

```
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size= 0.25, shuffle=True, random_state= 69)
```

```
x_train.shape
```

One hot encoding

```
print(y_train[:5])

y_train = keras.utils.to_categorical(y_train, num_classes = 2)

print(y_train[:5])
```

```
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```
model = Sequential()
```

```
model.add(Dense(units=32, activation='relu', input_shape = (x_train.shape[1],) ))
model.add(Dense(units=32, activation='relu'))
model.add(Dropout(0.50))
model.add(Dense(units=32, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(units=2, activation = 'softmax'))
```

```
model.compile(optimizer='Adam', loss= 'categorical_crossentropy', metrics = ['accuracy'])
model.summary()
```

```
model.fit(x_train, y_train, batch_size = 500, epochs=100, validation_split= 0.2, verbose = 2)
```

```
prediction = model.predict(x_test)
```

```
print(prediction[:5])
```

```
prediction = np.argmax(prediction, axis = 1)
print(prediction[:5])
```

```
from sklearn.metrics import accuracy_score, confusion_matrix
```

```
ac_score = accuracy_score(y_test, prediction)
print(ac_score)
```

```
con_matrix = confusion_matrix(y_test, prediction)
print(con_matrix)
```
